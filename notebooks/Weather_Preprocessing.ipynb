```python
df1 = pd.read_csv("/content/New York Queens 2024-01-01 to 2024-01-31.csv")
df2 = pd.read_csv("/content/New York Queens 2024-02-01 to 2024-02-29.csv")
df3 = pd.read_csv("/content/New york queens 2024-03-01 to 2024-03-31.csv")
df4 = pd.read_csv("/content/New york queens 2024-04-01 to 2024-04-30.csv")
df5 = pd.read_csv("/content/New York Queens 2024-05-01 to 2024-05-31.csv")
df6 = pd.read_csv("/content/New york queens 2024-06-01 to 2024-06-30.csv")
df7 = pd.read_csv("/content/New york queens 2024-07-01 to 2024-07-31.csv")
df8 = pd.read_csv("/content/New York Queens 2024-08-01 to 2024-08-31.csv")
df9 = pd.read_csv("/content/New York Queens 2024-09-01 to 2024-09-30.csv")
df10 = pd.read_csv("/content/New York Queens 2024-10-01 to 2024-10-31.csv")
df11 = pd.read_csv("/content/New york queens 2024-11-01 to 2024-11-30.csv")
df12 = pd.read_csv("/content/New York Queens 2024-12-01 to 2024-12-31.csv")
df13 = pd.read_csv("/content/New york queens 2025-01-01 to 2025-01-31.csv")
df14 = pd.read_csv("/content/New York Queens 2025-02-01 to 2025-02-28.csv")
df15 = pd.read_csv("/content/New York Queens 2025-03-01 to 2025-03-31.csv")
```

```python
frames = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12, df13, df14, df15]
df = pd.concat(frames)
```

```python
df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')
df['hour'] = df['datetime'].dt.hour
```

```python
for val in df['icon'].unique():
  print(f"{val}: {df[df['icon'] == val]['hour'].unique()}")
```

```python
def modify_icon(icon):
  icon = icon.replace("-day", "")
  icon = icon.replace("-night", "")
  icon = icon.replace("-", " ")

  return icon

df['icon'] = df['icon'].apply(modify_icon)
```

```python
print(df['icon'].unique())
```

```python
encoder = OneHotEncoder()
encoded_data = encoder.fit_transform(df[['icon']])
encoded_df = pd.DataFrame(encoded_data.toarray(), columns=encoder.get_feature_names_out(['icon']))
encoded_df
```

```python
print(encoded_df.value_counts())
print(df['icon'].value_counts())
```

```python
for col_name in encoded_df:
  df[col_name] = encoded_df[col_name]

df
```

```python
graphed_columns = ['Temperature', 'Feels Like', 'Dew', 'Humidity',
       'Precipitation', 'Snow Value', 'Snow Depth', 'Wind Gust', 'Wind Speed',
       'Wind Direction', 'Cloud Cover', 'Visibility', 'Solar Radiation',
       'Solar Energy', 'UV Index']

rows, cols = 5, 3
fig, axes = plt.subplots(rows, cols, figsize=(15, 20))

for i, col in enumerate(graphed_columns):
    row, column_index = int(i/3), i%3
    sns.histplot(df[col], kde=True, bins=30, ax=axes[row, column_index])
    axes[row, column_index].set_title(f"Histogram of {col}")

plt.tight_layout()
plt.show()
```
