```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import mean_squared_error, mean_absolute_error, median_absolute_error, r2_score, explained_variance_score
from tensorflow.keras.regularizers import l2

import numpy as np
import time
# --- Data Preparation ---
all_features = [
       'hour_sin','hour_cos', 'minute_sin', 'minute_cos', 'day_sin', 'day_cos',
       'week_num_sin', 'week_num_cos', 'month_sin', 'month_cos', 'is_holiday',
       'is_non_business_day', 'temp', 'feelslike', 'dew', 'humidity', 'precip',
       'snow', 'snowdepth', 'windspeed', 'winddir', 'cloudcover', 'visibility',
       'solarradiation', 'solarenergy', 'uvindex', 'hour','crash_count',
       'total_injuries', 'total_fatalities', 'crash_severity',
       'crash_occurred'
]
targets = ['speed', 'travel_time']

X = cleaned[all_features]
y = cleaned[targets]

# Use RobustScaler for better handling of outliers
scaler = RobustScaler()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False, random_state=42)
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

input_dim = X_train_scaled.shape[1]

# Build model
network = Sequential([
    Dense(256, input_dim=input_dim, activation='relu', kernel_regularizer=l2(0.001)),
    BatchNormalization(),
    Dropout(0.5),
    Dense(128, activation='relu', kernel_regularizer=l2(0.001)),
    BatchNormalization(),
    Dropout(0.4),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(2)
])



# Compile network
network.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae','r2_score'])

# Summary
network.summary()


history = network.fit(X_train_scaled, y_train,
                    validation_data=(X_test_scaled, y_test),
                    epochs=40,
                    batch_size=128)
```
